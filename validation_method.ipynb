# hold-out

```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, LeaveOneOut
from sklearn.linear_model import LinearRegression
df = sns.load_dataset('tips')
```

```python
df
```

```python
y_col = 'tip'
X = df.drop(columns=[y_col])
X
```

```python
numeric_cols = X.select_dtypes(include=np.number)
```

```python
numeric_cols
```

```python
X.dtypes
```

```python
#標準化のために数値カラムのリストを取得
numeric_cols = X.select_dtypes(include=np.number).columns.to_list()
```

```python
numeric_cols
```

```python
X = pd.get_dummies(X, drop_first=True)
X
```

```python
y = df[y_col]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train
```

```python
# 標準化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = X_train.copy()
X_train_scaled[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_train_scaled
```

```python
X_test_scaled = X_test.copy()
X_test_scaled[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
#　線形回帰モデル
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
y_pred
```

```python
X_test
```

```python
# 精度
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

## LOOCV

```python
#データ準備
X = df['total_bill'].values.reshape(-1, 1)
y = df['tip']
```

```python
loo =LeaveOneOut()
mse_list = []
model = LinearRegression()
for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    #モデルの学習
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)
```

```python
print(f"MSE(LOOCV):{np.mean(mse_list)}")
print(f"std:{np.std(mse_list)}")
```

```python
from sklearn.model_selection import cross_val_score
cv = LeaveOneOut()
scores = cross_val_score(model, X, y, cv=cv, scoring='neg_mean_squared_error')
print(f"MSE(LOOCV):{-np.mean(scores)}")
print(f"std:{np.std(scores)}")
```

## Pipeline

```python
from sklearn.pipeline import Pipeline
pipline = Pipeline(steps=[('scaler', StandardScaler()), ('model', LinearRegression())])
```
